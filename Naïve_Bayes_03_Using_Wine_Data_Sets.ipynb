# **`Naïve Bayes Using Wine Data Sets By (GaussianNB) & (MultinomialNB)`**

**Load `Sklearn` `Wine` Data Set :**

```python
from sklearn import datasets
wine = datasets.load_wine()
```

**ALL `Columns` Name :**

```python
dir(wine)
```

**ALL `feature_names`:**

```python
wine.feature_names
```

**ALL `target_names`:**

```python
wine.target_names
```

**Make a `pandas` `Data Frame` Through `wine` Data :**

```python
import pandas as pd

df = pd.DataFrame(wine.data,columns=wine.feature_names)
df.head(10)

# df.to_csv('Wine_Data_Sets.csv', index=False)
```

**Add `target` Column into the Data Frame :**

```python
df['target'] = wine.target
df.head(100)

# df.to_csv('Wine_Data_Sets.csv', index=False)
```

**`data` Looks Like :**

```python
wine.data[0:5]
```

**`Splite` the data for `training` & `tasting` purpose :**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(wine.data, wine.target, test_size=0.3, random_state=100)
```

**Make a `GaussianNB` model :**

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
```

**`Train` the model :**

```python
model.fit(X_train,y_train)
```

**`Accuracy` of the model :**

```python
model.score(X_test,y_test)
```

--------------------------------------------------------------

**Make a `MultinomialNB` Model :**

```python
from sklearn.naive_bayes import MultinomialNB

mn = MultinomialNB()
```

**`train` the Model :**

```python
mn.fit(X_train,y_train)
```

**`Accuracy` of the model :**

```python
mn.score(X_test,y_test)
```
