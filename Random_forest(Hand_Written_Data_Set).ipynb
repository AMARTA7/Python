**Make a Data Frame of *data & target* columns:**

```python
df = pd.DataFrame( digits.data,digits.target )
df.head()

# df.to_csv('Numbers.csv', index=False)
```

**Add *target* Column to the Data Frame:**

```python
df['target'] = digits.target
df.head()

# df.to_csv('Numbers.csv', index=False)
```

**Drop *targe*t column from the data set:**

```python
X = df.drop('target',axis='columns')
y = df.target
```

**Splite data for training and testing purpose (20%) :**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

**Size of train & testing Data:**

```python
print(f"Size of X_test Data: {len(X_test)}")
print(f"Size of X_train Data: {len(X_train)}")
```

**Create a Random Forest Classifier Model:**

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
```

**Train the Model through Splite Data:**

Hera we can change *'n_estimators' & 'criterion'* value. here default value-> *n_estimators=100 & criterion='gini'*

```python
model.fit(X_train, y_train)
```

**Accuracy of the model:**

```python
model.score(X_test, y_test)
```

**Confusion Matrix:**

```python
y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_predicted)
cm
```

**Better Visualisation of Confusion Matrix:**

```python
import matplotlib.pyplot as plt
import seaborn as sn

plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
