**Support Vector Machine (SVM) Using Python Sklearn Module (Iris Flower)**


*Load the Data:*

```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

*In data set ALL Columns name:*

```python
dir(iris)
```

**In data ALL Feature Names:**

```python
iris.feature_names
```

**In data ALL Target Names:**

```python
iris.target_names
```

*Print the Data Frame:*

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df.head(10)

# df.to_csv('Iris_Flower.csv', index=False)
```

Add *target* Column to the Data Frame:

```python
df['target'] = iris.target
df.head(10)
```

*Which Row Contain df.target==0:*

```python
df[df.target==0].head()
```

*Which Row Contain df.target==1:*

```python
df[df.target==1].head()
```

*Which Row Contain df.target==2:*

```python
df[df.target==2].head()
```

Make a new column ***fiower_name*** by the real Flower name. Like, ***0 == 'setosa' ; 1 == 'versicolor' ; 2 == 'virginica' ;*** For that use *lamda.*

```python
df['flower_name'] =df.target.apply(lambda x: iris.target_names[x])
df.head()

# df.to_csv('Iris_Flower.csv', index=False)
```

**versicolor** Flowers:

```python
df[df.target==1].head()
```

**virginica** Flowers:

```python
df[df.target==2].head()
```

Splite the data and make 3 data frame **df0, df1, df2.** w.r.to the Flower name:

```python
df0 = df[:50]
df1 = df[50:100]
df2 = df[100:]
```

**Sepal length (cm) vs Sepal Width (cm ) {(Setosa vs Versicolor)}**

```python
import matplotlib.pyplot as plt

plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'],color="green",marker='+')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'],color="red",marker='*')
```

**petal length (cm) vs	petal width (cm) {(Setosa vs Versicolor)}:**

```python
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'],color="green",marker='+')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'],color="red",marker='*')
```

**sepal length (cm) vs sepal width (cm) {(Setosa vs virginica)}:**

```python
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'],color="blue",marker='^')
plt.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'],color="magenta",marker='v')
```

**petal length (cm) vs petal width (cm) {(Setosa vs virginica)}:**

```python
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'],color="blue",marker='^')
plt.scatter(df2['petal length (cm)'], df2['petal width (cm)'],color="magenta",marker='v')
```

Prepare **X & y** for training and tasting Purposes:

```python
X = df.drop(['target','flower_name'], axis='columns')
y = df.target
```


*Splite data for training and testing purpose (20%) :*

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

*Size of train & testing Data:*

```python
print(f"Length of X_test Data: {len(X_test)}")
print(f"Length of X_train Data: {len(X_train)}")
```

**Create Support Vector Machine (SVM) Model:**

```python
from sklearn.svm import SVC
model = SVC()
```


***Train the Model through Splite Data:***

```python
model.fit(X_train, y_train)
```

**TEST-01 / (Prediction-01):**

```python
ans = model.predict([['5.1', '3.5', '1.4', '0.2']])
if ans == 0:
  print(f"This is 'Setosa Flower'!")
if ans == 1:
  print(f"'This is 'Versicolor Flower'!")
if ans == 2:
  print(f"'This is 'Virginica Flower'!")
```

**TEST-02 / (Prediction-02):**

```python
ans = model.predict([['7.0',	'3.2',	'4.7',	'1.4']])
if ans == 0:
  print(f"This is 'Setosa Flower'!")
if ans == 1:
  print(f"'This is 'Versicolor Flower'!")
if ans == 2:
  print(f"'This is 'Virginica Flower'!")
```

**TEST-03 / (Prediction-03):**

```python
ans = model.predict([['6.3',	'3.3',	'6.0',	'2.5']])
if ans == 0:
  print(f"This is 'Setosa Flower'!")
if ans == 1:
  print(f"'This is 'Versicolor Flower'!")
if ans == 2:
  print(f"'This is 'Virginica Flower'!")
```

**Accuracy of the model:**

```python
model.score(X_test, y_test)
```
