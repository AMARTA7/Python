**Support Vector Machine (SVM) Using Python Sklearn Module (Hand Written Data Set)**

**Make a Data Frame of *data & target* columns:**

```python
df = pd.DataFrame( digits.data,digits.target )
df.head()

# df.to_csv('Numbers.csv', index=False)
```

**Add *target* Column to the Data Frame:**

```python
df['target'] = digits.target
df.head()

# df.to_csv('Numbers.csv', index=False)
```

**Splite data for training and testing purpose (30%) :**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('target',axis='columns'), df.target, test_size=0.3)
```

**Size of train & testing Data:**

```python
print(f"Size of X_test Data: {len(X_test)}")
print(f"Size of X_train Data: {len(X_train)}")
```

**Create Support Vector Machine (SVM) Model:**

```python
from sklearn.svm import SVC

rbf_model = SVC(kernel='rbf')
```

**Train the Model through Splite Data:**

```python
rbf_model.fit(X_train, y_train)
```

**Accuracy of the model (rbf_model):**

```python
rbf_model.score(X_test,y_test)
```

**## Using Linear kernel Model:**

```python
linear_model = SVC(kernel='linear')

linear_model.fit(X_train,y_train)
```

**Accuracy of the model (linear_model):**

```python
linear_model.score(X_test,y_test)
```
